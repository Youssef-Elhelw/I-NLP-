# Natural Language Processing (NLP) is a field of artificial intelligence that focuses on enabling computers to understand, interpret, and generate human language. In Python, two of the most widely used NLP libraries are SpaCy and NLTK (Natural Language Toolkit).

## 🟢 What is SpaCy 

- SpaCy is a modern NLP library designed for speed and production use.

Pros 💚:

- Support for 75+ languages
- 84 trained pipelines for 25 languages
- Multi-task learning with pretrained transformers like BERT
- Pretrained word vectors
- State-of-the-art speed
- Production-ready training system
- Linguistically-motivated tokenization
- Components for named entity recognition, part-of-speech tagging, dependency parsing, sentence segmentation, text classification, lemmatization, morphological analysis, entity linking and more
- Easily extensible with custom components and attributes
- Support for custom models in PyTorch, TensorFlow and other frameworks
- Built in visualizers for syntax and NER
- Easy model packaging, deployment and workflow management
- Robust, rigorously evaluated accuracy

- Supports many NLP tasks:

    - Tokenization
    - Stemming & Lemmatization
    - POS tagging
    - Parsing

```python
import spacy

# Load the English model
nlp = spacy.load("en_core_web_sm") # now 'nlp' is an object that we can use to process text with SpaCy.

# Example text
text = ("Apple is looking at buying U.K. startup for $1 billion. but the deal is not finalized yet. The CEO of Apple, Tim Cook, said that they are excited about the potential acquisition. However, some analysts are skeptical about the deal and think that it might not go through.")

doc = nlp(text) # 'doc' is a SpaCy object that contains the processed text and allows us to access various linguistic features.
```

### ⚫ Tokenization

```python

for sent in doc.sents:
    print(sent.text)
    print("Tokens:")
    for token in sent:
        print(f"{token}")
```

## 🟢 What is NLTK

- NLTK is one of the oldest and most popular NLP libraries in Python. It is designed mainly for education and research.

NLTK is a leading platform for building Python programs to work with human language data.

Pros 💚 
- It provides easy-to-use interfaces to over 50 corpora and lexical resources such as WordNet.
- along with a suite of text processing libraries for 

    - classification
    - tokenization
    - stemming
    - tagging
    - parsing
    - and semantic reasoning

```python
import nltk
nltk.download('punkt') # Download the necessary resources for tokenization
from nltk.tokenize import sent_tokenize, word_tokenize
```

### ⚫ Tokenization

```python
sentences = sent_tokenize(text)
for sent in sentences:
    print(sent)
    print("Tokens:")
    tokens = word_tokenize(sent)
    for token in tokens:
        print(f"{token}")
```
